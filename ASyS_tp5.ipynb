##Grfico de funciones
Se muestra a continuacin un formato general para definir y graficar funciones tanto en tiempo discreto como en tiempo continuo.



###Tiempo discreto(CLASE)

````python
# Definicin del vector de tiempo discreto

dn = 1
n_min = -10
n_max = 10

n = np.arange(n_min, n_max + dn, dn)

#Definicin de seal de ejemplo a graficar

x = 2*n

#Grfica de la funcin

plt.figure(figsize=[10,5])
plt.axhline(0, color='gray',linewidth=1.2)
plt.axvline(0, color='gray',linewidth=1.2)
plt.grid()

plt.plot(n, x, '.b', label='$x[n]=2*n$')
plt.title('Seal de ejemplo: $x[n]=2*n$')
plt.xlim([n_min,n_max])
plt.xlabel('n')

plt.legend()
````

###Tiempo continuo(CLASE)

````python
# Definicin del vector de tiempo continuo

dt = 0.01     
t_min = -5
t_max = 5

t = np.arange(t_min, t_max + dt, dt)

#Definicin de seal de ejemplo a graficar

x = 2*t

#Grfica de la funcin

plt.figure(figsize=[10,5])
plt.axhline(0, color='gray',linewidth=1.2)
plt.axvline(0, color='gray',linewidth=1.2)
plt.grid()

plt.plot(t, x, 'b', label='$x(t)=2*t$')
plt.title('Seal de ejemplo: $x(t)=2*t$')
plt.xlim([t_min,t_max])
plt.xlabel('t')

plt.legend()
````

##Funcin escaln:
Para graficar el escaln unitario, se utilizar la funcin *np.heaviside*. 

A continuacin se muestra un ejemplo del formato general para armar y graficar funciones, y en especfico para el escaln unitario:


###Escaln en tiempo discreto(CLASE)

````python
# Definicin del vector de tiempo discreto

dn = 1
n_min = -10
n_max = 10

n = np.arange(n_min, n_max + dn, dn)

#Definicin de seal escaln a graficar

u_n = np.heaviside(n,1)

#Grfica de la funcin

plt.figure(figsize=[10,5])
plt.axhline(0, color='gray',linewidth=1.2)
plt.axvline(0, color='gray',linewidth=1.2)
plt.grid()

plt.plot(n,u_n,'.b',label='$u[n]$')
plt.title('Escaln unitario en tiempo discreto: $u[n]$')
plt.xlim([n_min,n_max])
plt.xlabel('n')

plt.legend()
````

###Escaln en tiempo continuo(CLASE)

````python
# Definicin del vector de tiempo continuo

dt = 0.01     # Time step
t_min = -5
t_max = 5

t = np.arange(t_min, t_max + dt, dt)

#Definicin de seal escaln

u_t = np.heaviside(t,1)

#Grfica de la funcin

plt.figure(figsize=[10,5])
plt.axhline(0, color='gray',linewidth=1.2)
plt.axvline(0, color='gray',linewidth=1.2)
plt.grid()

plt.plot(t, u_t, 'b', label='$u(t)$')
plt.title('Escaln unitario en tiempo continuo: $u(t)$')
plt.xlim([t_min,t_max])
plt.xlabel('t')

plt.legend()
````

# Ejercicio N7(CLASE)
# Transformaciones de la variable independiente

Se ejemplifica en el siguiente ejercicio una manera de visualizar las distintas transformaciones que se pueden aplicar a la variable independiente (desplazamiento, inversin y escalado en el tiempo).

Se utiliz el ejercicio 7 del prctico de gabinete visto en clase.

````python
# Definicin del vector de tiempo continuo

dt = 0.01     
t_min = -7
t_max = 7

t = np.arange(t_min, t_max + dt, dt)

def plot(a,b):

#Definicin de seal

  xa = (t+1)*np.heaviside(t+1,1)*np.heaviside(-t+1,1)
  xb = 2*np.heaviside(t-1,1)*np.heaviside(-t+2,1)

  x = xa + xb

  ti=t*a+b

  xai = (ti+1)*np.heaviside(ti+1,1)*np.heaviside(-ti+1,1)
  xbi = 2*np.heaviside(ti-1,1)*np.heaviside(-ti+2,1)
  xi = xai + xbi

#Grfica de la funcin

  plt.figure(figsize=[10,5])
  plt.axhline(0, color='gray',linewidth=1.2)
  plt.axvline(0, color='gray',linewidth=1.2)
  plt.grid()

  plt.plot(t, x, 'b', label='$x(t)$')
  plt.plot(t, xi, 'r', label='$x(a*t+b)$')
  plt.title('Ejemplo de transformacin de la variable independiente')
  plt.xlim([t_min,t_max])
  plt.ylim(-5, 5)
  plt.xlabel('t')

  plt.legend()

ipwidgets.interact(plot, a=(-2, 2, 0.5), b=(-5,5,0.1))
````

# Ejercicio N9(CLASE)

Se muestra un ejercicio modelo para visualizar variaciones en amplitud y frecuencia de una senoidal pura.

````python
# Definicin del vector de tiempo continuo

dt = 0.01     
t_min = -7
t_max = 7

t = np.arange(t_min, t_max + dt, dt)

def plot(wa,A):

#Definicin de seal

  xa = A*np.sin(wa*t)

#Grfica de la funcin

  plt.figure(figsize=[10,5])
  plt.axhline(0, color='gray',linewidth=1.2)
  plt.axvline(0, color='gray',linewidth=1.2)
  plt.grid()

  plt.plot(t, xa, 'b', label='$x(t)=A*sin(wa*t)$')
  plt.title('Seal senoidal con amplitud y frecuencia variable')
  plt.xlim([t_min,t_max])
  plt.ylim(-5, 5)
  plt.xlabel('t')

  plt.legend()

ipwidgets.interact(plot, wa=(0, 20, 0.1), A=(-5,5,0.5))
````


# Ejercicio N10(CLASE)

Se visualiza a continuacin una seal compuesta por dos seales senoidales.

Variar las frecuencias y amplitudes para visualizar distintas formas de onda.

Pregunta: Para qu valores de los sliders la seal resultante deja de ser peridica?


````python
# Definicin del vector de tiempo continuo

dt = 0.01     
t_min = -20
t_max = 20

t = np.arange(t_min, t_max + dt, dt)

def plot(w_xa, amp_xa, w_xb, amp_xb):

#Definicin de seal

  xa = amp_xa*np.sin(w_xa*t)
  xb = amp_xb*np.sin(w_xb*t)
  x = xa + xb

#Grfica de la funcin

  plt.figure(figsize=[20,5])
  plt.axhline(0, color='gray',linewidth=1.2)
  plt.axvline(0, color='gray',linewidth=1.2)
  plt.grid()

  plt.plot(t, xa, 'y', label='$xa(t)$')
  plt.plot(t, xb, 'r', label='$xb(t)$')
  plt.plot(t, x, 'b', label='$x(t)=xa(t)+xb(t)$')
#  plt.plot(t, xi, 'r', label='$xi$')
  plt.title('Sumatoria de dos senoidales')
  plt.xlim([t_min,t_max])
  plt.ylim(-5, 5)
  plt.xlabel('t')

  plt.legend()

ipwidgets.interact(plot, w_xa=(0, 20, 0.01), amp_xa=(-4,5,0.5),w_xb=(0, 20, 0.01), amp_xb=(-4,5,0.5))
````
