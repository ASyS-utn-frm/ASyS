```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
from matplotlib.animation import FuncAnimation
from matplotlib import rc
import matplotlib.ticker as ticker
from scipy import signal
plt.rcParams['figure.figsize'] = [10, 5]
```

EJERCICIO 2020.09.20:

Una señal de banda limitada y sobremuestreada presenta el espectro de frecuencias que se muestra en la figura. Si el intervalo de muestreo utilizado es $\Delta t=120\,\mu s$, determinar:

a) La frecuencia lineal de muestreo $f_s$

b) La frecuencia angular máxima de la señal $\omega_{max}$

c) La frecuencia máxima que puede muestrearse correctamente con la $f_s$ utilizada.

d) ¿Podría haberse muestreado la misma señal con una frecuencia de muestreo menor? ¿Cuál es la mínima posible?

```python
def triang(x, a, τ):
    return [a*max(0, e) for e in -signal.sawtooth(τ * np.pi * x, 0.5)]

f, ax = plt.subplots(1)
x = np.linspace(-3, 3, 400)
ax.plot(x, triang(x, a=1, τ=1))
plt.xlabel('$\Omega$')
plt.grid()
plt.title('|X(j$\Omega$)|')
ax.xaxis.set_major_formatter(ticker.FormatStrFormatter('%g $\\pi$'))
ax.xaxis.set_major_locator(matplotlib.ticker.MultipleLocator(base=.5))
plt.show()
```

EJERCICIO 2020.09.23:

Una señal de banda limitada con espectro de forma triangular es submuestreada. El espectro de la señal de tiempo discreto resultante se muestra en la figura. Si se sabe que la frecuencia de alias es $f_{alias}=2\,kHz$, determinar:

a) La frecuencia angular de muestreo $w_s$

b) La frecuencia angular máxima de la señal $\omega_{max}$

c) ¿Que frecuencia de muestreo debería haberse utilizado para muestrear correctamente la señal?

```python
def triang(x, a, τ):
    return np.array([a*max(0, e) for e in -signal.sawtooth(τ * np.pi * x, 0.5)])


per = 1/3
f, ax = plt.subplots(1)
x = np.linspace(-1.5, 1.5, 400)
x1 = np.linspace(-0.5, 0.5, 400)
x2 = np.linspace(-0.5, 0.5, 400)
x3 = np.linspace(-0.5, 0.5, 400)
x4 = np.linspace(0.5, 1.5, 400)
x5 = np.linspace(-1.5, 0.5, 400)
ax.plot(x1, triang(x1, a=1, τ=per),'b')
ax.plot(x3+2, triang(x3, a=1, τ=per),'b')
ax.plot(x2-2, triang(x2, a=1, τ=per),'b')
ax.plot(x, triang(x, a=1, τ=per),':b')
ax.plot(x5+2, triang(x5, a=1, τ=per),':b')
ax.plot(x4-2, triang(x4, a=1, τ=per),':b')
ax.plot(x4, triang(x4, a=1, τ=per)+triang(x4-2, a=1, τ=per),'b')
ax.plot(x5, triang(x5, a=1, τ=per)+triang(x5+2, a=1, τ=per),'b')
plt.xlabel('$\Omega$')
plt.grid()
plt.title('|X(j$\Omega$)|')
plt.xlim([-2.5,2.5])
plt.ylim([0,1.1])
ax.xaxis.set_major_formatter(ticker.FormatStrFormatter('%g $\\pi$'))
ax.xaxis.set_major_locator(matplotlib.ticker.MultipleLocator(base=.5))
plt.show()
```
